Compare the ranking of countries by Composite Score with their HDI ranking and discuss any

```python
def find_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[col] < lower) | (data[col] > upper)]

# Find outliers
hdi_outliers = find_outliers(south_asia, 'hdi')
gni_outliers = find_outliers(south_asia, 'gross_inc_percap')

print("HDI outliers:")
print(hdi_outliers[['country', 'hdi']])

print("\nGNI outliers:")
print(gni_outliers[['country', 'gross_inc_percap']])

# Combine all outliers
outliers = pd.concat([hdi_outliers, gni_outliers]).drop_duplicates()

# Create scatter plot
plt.figure(figsize=(10, 6))

# Plot all points with HDI-based color mapping
scatter = plt.scatter(
    south_asia['gross_inc_percap'],
    south_asia['hdi'],
    c=south_asia['hdi'],
    cmap='plasma',        # changed colormap (unique & clear)
    alpha=0.7,
    label='Countries'
)

# Overlay outliers with black edge
plt.scatter(
    outliers['gross_inc_percap'],
    outliers['hdi'],
    facecolors='none',
    edgecolors='black',
    s=120,
    linewidth=1.5,
    label='Outliers'
)

# Colorbar
plt.colorbar(scatter, label='HDI Value')

# Labels and title
plt.title('GNI per Capita vs HDI (Outliers Highlighted)')
plt.xlabel('GNI per Capita')
plt.ylabel('HDI')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

7. Outlier Detection:

• Detect outliers in HDI {"hdi"} and GNI per Capita {"gross inc percap"} for each region using
the 1.5 × IQR rule.

```python
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    return df[(df[column] < Q1 - 1.5*IQR) | (df[column] > Q3 + 1.5*IQR)]

sa_outliers = detect_outliers(sa_2020_22, 'hdi')
me_outliers = detect_outliers(me_2020_22, 'hdi')
```


• Create scatter plots highlighting outliers in a different color.

```python
sns.scatterplot(data=sa_2020_22, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=sa_outliers, x='gross_inc_percap', y='hdi', color='red')
plt.title("South Asia HDI Outliers")
plt.show()
```

```python
sns.scatterplot(data=me_2020_22, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=me_outliers, x='gross_inc_percap', y='hdi', color='red')
plt.title("Middle East HDI Outliers")
plt.show()
```
